```python
# load libraries, use "dl" conda env on wsl2, has tensorflow
# python v 3.8

import numpy as np
from tensorflow.keras.applications import resnet50
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.applications.resnet50 import \
    decode_predictions
import matplotlib.pyplot as plt
import tensorflow as tf
#tf.get_logger().setLevel()

%matplotlib inline 
```

```python
# sample images saved to local directory, 
# relative path: ./data/dogs/dog1.jpg and dog2.jpg  
# relative path: ./data/cats/cat1.jpg and cat2.jpg  
# manually resized internet download images to 224x224 pixels.  aspect ratio maintained. colors rgb.
```

```python
# Load Image, test pictures from local repo  
image = load_img('./data/cats/cat1.jpg', target_size=(224, 224))
image
```

```python
# All worked!  Next use numpy to convert image pixels to matrix of numbers, 224x224, 3 color channels, n-images. 
# 4-D tensors
image_np = img_to_array(image)
image_np = np.expand_dims(image_np, axis=0)
image_np
```

```python
image = load_img('./data/dogs/dog1.jpg', target_size=(224, 224))
# 4-D tensors, expand dog1 image
image_np = img_to_array(image)
image_np = np.expand_dims(image_np, axis=0)
image_np
```
